```python
import pandas as pd
```

```python
customer_master = pd.read_csv('data/practice01/customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('data/practice01/item_master.csv')
item_master.head()
```

```python
transaction_1 = pd.read_csv('data/practice01/transaction_1.csv')
transaction_1.head()
```

```python
transaction_detail_1 = pd.read_csv('data/practice01/transaction_detail_1.csv')
transaction_detail_1.head()
```

```python
transaction_2 = pd.read_csv('data/practice01/transaction_2.csv')
transaction_2.head()
```

```python
# Concatenate multiple data (Union - Vertical)
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
transaction_detail_2 = pd.read_csv('data/practice01/transaction_detail_2.csv')
transaction_detail_2.head()
transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index=True)
transaction_detail.head()
```

```python
print(len(transaction_detail_1))
print(len(transaction_detail_2))
print(len(transaction_detail))
```

```python
# Merge multiple data (Join - Horizontal)
join_data = pd.merge(transaction_detail, transaction[['transaction_id', 'payment_date', 'customer_id']], 
                     on='transaction_id', how='left')
join_data.head()
```

```python
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

```python
join_data = pd.merge(join_data, customer_master, on='customer_id', how='left')
join_data = pd.merge(join_data, item_master, on='item_id', how='left')
join_data.head()
```

```python
join_data['price'] = join_data['quantity'] * join_data['item_price']
join_data[['quantity', 'item_price', 'price']].head()
```

```python
join_data['price'].sum() == transaction['price'].sum()
```

```python
# Check null value counts
join_data.isnull().sum()
```

```python
join_data.describe()
```

```python
print(join_data['payment_date'].min())
print(join_data['payment_date'].max())
```

```python
# Check the data type for each column
join_data.dtypes
```

```python
# Convert object data to datetime data
join_data['payment_date'] = pd.to_datetime(join_data['payment_date'])
join_data['payment_month'] = join_data['payment_date'].dt.strftime('%Y%m')
join_data[['payment_date', 'payment_month']].head()
```

```python
join_data.groupby('payment_month').sum()['price']
```

```python
join_data.groupby(['payment_month', 'item_name']).sum()[['price', 'quantity']]
```

```python
pd.pivot_table(join_data, index='item_name', columns='payment_month', 
               values=['price', 'quantity'], aggfunc='sum')
```

```python
graph_data = pd.pivot_table(join_data, index='payment_month', columns='item_name',
                           values='price', aggfunc='sum')
graph_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
plt.plot(list(graph_data.index), graph_data['PC-A'], label='PC-A')
plt.plot(list(graph_data.index), graph_data['PC-B'], label='PC-B')
plt.plot(list(graph_data.index), graph_data['PC-C'], label='PC-C')
plt.plot(list(graph_data.index), graph_data['PC-D'], label='PC-D')
plt.plot(list(graph_data.index), graph_data['PC-E'], label='PC-E')
plt.legend()
```
